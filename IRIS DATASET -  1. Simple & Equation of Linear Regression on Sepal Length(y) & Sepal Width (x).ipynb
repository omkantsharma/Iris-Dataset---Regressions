## Simple Linear Regression

### Create a Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()  #use alias
```

### Train the Model

```python
lr.fit(x_train,y_train)  # fit the algorithm through method
```

```python
# R2_score - goodness of fit on train data set
lr.score(x_train,y_train)
```

```python
# r2_score - goodness of fit on test data set
lr.score(x_test,y_test)
```

### Make Predictions

#### Method -1

```python
y_pred=lr.predict(x_test) # use x_test for prediction of y 
# (here came predicted y)
```

```python
print(y_pred)
print(y_pred.shape)
print(type(y_pred))
```

### Equation of Linear Regression

#### Train the Model

```python
#  m = slope/coeffieient or a 
lr.coef_         #  a or m
```

### Evaluation - simple Linear Regression

```python
import numpy as np
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
```

```python
mabe = mae(y_test,y_pred)
msqe=mse(y_test,y_pred)
rmsqe=np.sqrt(mse(y_test,y_pred)) 
r2sc=r2(y_test,y_pred)
```

```python
print(f"Simple Linear Regression's Mean Absolute Error is - {msqe} ")
print(f"Simple Linear Regression's Mean Square Error is - {msqe} ")
print(f"Simple Linear Regression's Root Mean Square Error is - {rmsqe} ")
print(f"Simple Linear Regression's Root Squared Score is - {r2sc} ")
```
